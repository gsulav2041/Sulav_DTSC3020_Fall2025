## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here

# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Create scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',      # Color by species
    palette='Set1',     # Color palette
    s=80                # Marker size
)
plt.title('Penguin Bill Length vs Bill Depth by Species')
```

**Your interpretation (2–3 sentences):**

From the scatter pplot it shows that different penguin species have distinct bill shapes. Adelie penguins have generally shorter bills, Gentoo penguins have longer and deeper bills and chinstrap penguins fall in between those two.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins

# Import libraries (if not already imported)
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Histogram with 10 bins
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.hist(penguins['bill_length_mm'].dropna(), bins=10, color='skyblue', edgecolor='black')
plt.title('Bill Length Histogram (10 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')
```

```python
# TODO: Histogram with 30 bins
# Histogram with 30 bins
plt.subplot(1, 2, 2)
plt.hist(penguins['bill_length_mm'].dropna(), bins=30, color='salmon', edgecolor='black')
plt.title('Bill Length Histogram (30 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

**Your comparison (1–2 sentences):**


On the 10 bins, the histogram has broad bars that is the overview of the distribution. On the 30 bins the bars are thin and the higher peaks stating the number going up. The plot appears to be noisy and hard to interpret.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species

# Import libraries (if not already imported)
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Create a boxplot of body mass by species
plt.figure(figsize=(8,6))
sns.boxplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    palette='Set2'
)
plt.title('Penguin Body Mass by Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()

```

**Your summary (1–2 sentences):**

On the boxplot, it shows the Gentoo pengiuns to be genereally heavier than other two. Chinstrap being in the middle adn Adelie being the lighted of them all.
